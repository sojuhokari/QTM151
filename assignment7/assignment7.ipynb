(c) Clean a column

<font size = "3">

- Import the dataset "data_raw/results.csv" to a new object "results"
- Our goal is to replace any string values in the column "milliseconds" <br>
to missing values (NaNs) and covert to numeric
- To do so
    - "Extract list of non-numeric values"
    - "Replace certain values"
    - "Convert column to numeric"
    - "Display"
 


```python
results = pd.read_csv("data_raw/results.csv")

# Find the unique non-numeric values
print(
    pd.unique(
        results.query("milliseconds.str.isnumeric() == False")["milliseconds"]
    )
)

# There is only one unique non-numeric value, \N, so we replace it.
results["milliseconds"] = results["milliseconds"].replace("\\N", np.nan)

# Convert the column to numeric
results["milliseconds"] = pd.to_numeric(results["milliseconds"])

# Display
display(results["milliseconds"])
```

(d) Groupby + Aggregate

<font size = "3">

- Compute the mean and standard deviation of "laps", <br>
grouping by "constructorId". Store the new dataset <br>
as "df_aggCon_pos"
- Sort the aggregate dataset in decending <br>
order of mean laps using ".sort_values()"



```python
# Groupby and aggregate
df_aggCon_laps = (results
    .groupby("constructorId")
    .agg(
        laps_mean = ("laps", "mean"),
        laps_sd = ("laps", "std")
    )
)

# Sort values
df_aggCon_laps = df_aggCon_laps.sort_values(
    by = "laps_mean",
    ascending = False
)

display(df_aggCon_laps)
```

(e) Query + Groupby + Aggregate

<font size = "3">

- Use ".query()" to subset observations with <br>
'race_id >= 150'
- Compute the mean and standard deviation of "position", <br>
grouping by "constructorId". Store the new dataset <br>
as "df_subAggCon_pos"


```python
# Clean up position; convert to numeric
results["position"] = results["position"].replace("\\N", np.nan)
results["position"] = pd.to_numeric(results["position"])

# Do the query, groupby, and aggregation
df_subAggCon_pos = (results
    .query("raceId >= 150")
    .groupby("constructorId")
    .agg(
        position_mean = ("position", "mean"),
        position_sd = ("position", "std")
    )
)

display(df_subAggCon_pos)
```
